```python
from sklearn import datasets 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import svm 
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
#load digits data set
digits = datasets.load_digits()
#Feature and tragets array
X = digits.data
y = digits.target
#split to training and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#SVM
#Create a svm Classifier with 3 different kernels
linear = svm.SVC(kernel='linear')
rbf = svm.SVC(kernel='rbf')
sigmoid = svm.SVC(kernel='sigmoid')
#Train the model using the training sets
linear.fit(X_train, y_train)
rbf.fit(X_train, y_train)
sigmoid.fit(X_train, y_train)
#Predict the response for test dataset
y_pred_lin = linear.predict(X_test)
y_pred_rbf = rbf.predict(X_test)
y_pred_sig = sigmoid.predict(X_test)
#print metrics classification
print(metrics.classification_report(y_test, y_pred_lin, labels=[1, 2, 3, 4, 5, 6, 7, 8, 9]))
print(metrics.classification_report(y_test, y_pred_rbf, labels=[1, 2, 3, 4, 5, 6, 7, 8, 9]))
print(metrics.classification_report(y_test, y_pred_sig, labels=[1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
#The rbf kernel scored highest and will be used as the SVM score
#save score for svm
svm_score = metrics.accuracy_score(y_test, y_pred_rbf)
```

```python
#Naïve Bayes
#Create NB classifier
nb = MultinomialNB()
#Fit Data
nb.fit(X_train, y_train)
#Predict
y_pred = nb.predict(X_test)
#save score
nb_score = metrics.accuracy_score(y_test, y_pred)
confusion_matrix(y_test, y_pred)
```
